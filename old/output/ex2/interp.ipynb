```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
import glob
for path in glob.glob("*.pkl"):
    with open(path, 'rb') as f:
        data = pkl.load(f)
    print(path)
    front_to_back = np.array([x['fc0_fc1']['val_acc'] for x in data])
    back_to_front = np.array([x['fc1_fc0']['val_acc'] for x in data])
    # plt.plot(range(len(front_to_back)), front_to_back, label='fc0_fc1')
    # plt.plot(range(len(back_to_front)), back_to_front, label='fc0_fc1')
    # plt.legend()
    plt.plot(range(len(front_to_back)), (front_to_back - back_to_front), label='fc0_fc1')
    # plt.plot(range(len(back_to_front)), back_to_front, label='fc0_fc1')
    # plt.legend()
    plt.show()
```
