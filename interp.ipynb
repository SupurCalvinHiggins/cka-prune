```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
all_l1_vals = []
for seed in range(30):
    try:
        # f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-l1_module-fc1.pkl"
        with open(f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-l1_module-fc0-fc1.pkl", "rb") as f:
            output = pkl.load(f)
        l1_vals = [x["after_train"]["val_acc"] for x in output["output"]]
        print(l1_vals)
        all_l1_vals.append(l1_vals)
    except:
        pass
# all_l1_vals = all_l1_vals[1:]
all_l1_means = np.array(all_l1_vals).mean(axis=0)
all_l1_std = np.array(all_l1_vals).std(axis=0)
```

```python
all_cka_vals = []
for seed in range(30):
    try:
        # f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-cka_module-fc1.pkl"
        with open(f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-cka_module-fc0-fc1.pkl", "rb") as f:
            output = pkl.load(f)
        cka_vals = [x["after_train"]["val_acc"] for x in output["output"]]
        print(cka_vals)
        all_cka_vals.append(cka_vals)
    except:
        pass
# all_cka_vals = all_cka_vals[1:]
all_cka_means = np.array(all_cka_vals).mean(axis=0)
all_cka_std = np.array(all_cka_vals).std(axis=0)
```

```python
x = [0.2 * (0.8) ** (i - 1) for i in range(1, len(all_l1_means)+ 1)]
for i in range(1, len(x)):
    x[i] += x[i-1]
print(x)

plt.errorbar(x, all_l1_means, yerr=[all_l1_std, all_l1_std], label='l1')
plt.errorbar(x, all_cka_means, yerr=[all_cka_std, all_cka_std], label='cka')
plt.title("Iterative pruning with L1 and CKA")
plt.xlabel("Proportion of neurons in fc0 and fc1 pruned")
plt.ylabel("Validation accuracy")

plt.legend()
plt.show()
```
