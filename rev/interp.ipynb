```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
all_l1_vals = []
for seed in range(30):
    try:
        with open(f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-l1_module-fc1.pkl", "rb") as f:
            output = pkl.load(f)
        l1_vals = [x["after_train"]["val_acc"] for x in output["output"]]
        print(l1_vals)
        all_l1_vals.append(l1_vals)
    except:
        pass
# all_l1_vals = all_l1_vals[1:]
all_l1_means = np.array(all_l1_vals).mean(axis=0)
all_l1_std = np.array(all_l1_vals).std(axis=0)
```

```python
all_cka_vals = []
for seed in range(30):
    try:
        with open(f"output/pruned-lenet-300-100_seed-{seed}_p-0.2_type-cka_module-fc1.pkl", "rb") as f:
            output = pkl.load(f)
        cka_vals = [x["after_train"]["val_acc"] for x in output["output"]]
        print(cka_vals)
        all_cka_vals.append(cka_vals)
    except:
        pass
# all_cka_vals = all_cka_vals[1:]
all_cka_means = np.array(all_cka_vals).mean(axis=0)
all_cka_std = np.array(all_cka_vals).std(axis=0)
```

```python
plt.errorbar(range(len(all_l1_means)), all_l1_means, yerr=[all_l1_std, all_l1_std], label='l1')
plt.errorbar(range(len(all_cka_means)), all_cka_means, yerr=[all_cka_std, all_cka_std], label='cka')
plt.legend()
plt.show()
```
