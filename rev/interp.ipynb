```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open("output/pruned-lenet-300-100_seed-0_p-0.2_type-l1_module-fc1.pkl", "rb") as f:
    output = pkl.load(f)
l1_vals = [x["after_train"]["val_acc"] for x in output["output"]]
```

```python
with open("output/pruned-lenet-300-100_seed-0_p-0.2_type-cka_module-fc1.pkl", "rb") as f:
    output = pkl.load(f)
cka_vals = [x["after_train"]["val_acc"] for x in output["output"]]
```

```python
plt.plot(range(len(l1_vals)), l1_vals, label='l1')
plt.plot(range(len(cka_vals)), cka_vals, label='cka')
plt.legend()
plt.show()
```
